```python
'''  '''

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def mandel_bool(z0, c, n):   #  ,  ,         
    zn = z0
    i = 0
    while abs(zn) < 2 and i < n:
        i+=1
        zn = zn**2 + c 
    
    return i == n
    

x = np.linspace(start=-1.5, stop=0.6, num=1000)
y = np.linspace(start=-1, stop=1, num=1000)

lz = []
for i in x:
    for j in y:
        lz.append(complex(i,j))

z = np.array(lz)                   #    

lz_mand = []                       #      
for i in z:                        #
    if mandel_bool(0,i,100):       #
        lz_mand.append(i)          #
                                   #
z_mand = np.array(lz_mand)         # 


fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.scatter(x = z_mand.real, y = z_mand.imag, marker = '.')
axes.axis('equal')
```
